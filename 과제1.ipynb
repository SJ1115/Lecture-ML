```python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE, ADASYN
from plotly.subplots import make_subplots
import plotly.graph_objects as go
X, y = make_classification(n_classes=3, weights=[0.03, 0.07,0.9],n_features=2,
    n_redundant=0, n_clusters_per_class=1, n_samples=1000, random_state=0)
print('Original dataset shape %s' % Counter(y)) 
```

```python
sm0 = SMOTE(k_neighbors=3)
X0, y0 = sm0.fit_resample(X, y)

sm1 = SMOTE(k_neighbors=6)
X1, y1 = sm1.fit_resample(X, y)

sm2 = SMOTE(k_neighbors=9)
X2, y2 = sm2.fit_resample(X, y)

print('Original dataset shape %s' % Counter(y1)) 
```

```python
def SetColor(x):
    if x==0:
        return 'red'
    elif x==1:
        return 'green'
    else :
        return 'blue'
```

```python
fig = make_subplots(
    rows=2, cols=2,
    column_widths=[2, 2], row_heights=[2, 2],
    subplot_titles=("Original", "SMOTE with k=3", "SMOTE with k=6", "Smote with k=9"))

fig.add_trace(
    go.Scatter(x=X[:,0], y=X[:,1],
        mode='markers', marker=dict(color=list(map(SetColor, y))),
        showlegend=False),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=X0[:,0], y=X0[:,1],
              mode='markers', marker=dict(color=list(map(SetColor, y0))),
              showlegend=False),
    row=1, col=2
)

fig.add_trace(
    go.Scatter(x=X1[:,0], y=X1[:,1],
              mode='markers', marker=dict(color=list(map(SetColor, y1))),
              showlegend=False),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(x=X2[:,0], y=X2[:,1],
              mode='markers', marker=dict(color=list(map(SetColor, y2))),
              showlegend=False),
    row=2, col=2
)

fig.update_layout(title_text='SMOTE with K_neighbors')

fig.show()
```

```python
ad0 = ADASYN(n_neighbors=3)
X3, y3 = ad0.fit_resample(X, y)

ad1 = ADASYN(n_neighbors=6)
X4, y4 = ad1.fit_resample(X, y)

ad2 = ADASYN(n_neighbors=9)
X5, y5 = ad2.fit_resample(X, y)

print('Original dataset shape %s' % Counter(y1)) 
```

```python
fig = make_subplots(
    rows=2, cols=2,
    column_widths=[2, 2], row_heights=[2, 2],
    subplot_titles=("Original", "ADASYN with k=3", "ADASYN with k=6", "ADASYN with k=9"))

fig.add_trace(
    go.Scatter(x=X[:,0], y=X[:,1],
        mode='markers', marker=dict(color=list(map(SetColor, y))),
        showlegend=False),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=X3[:,0], y=X3[:,1],
              mode='markers', marker=dict(color=list(map(SetColor, y3))),
              showlegend=False),
    row=1, col=2
)

fig.add_trace(
    go.Scatter(x=X4[:,0], y=X4[:,1],
              mode='markers', marker=dict(color=list(map(SetColor, y4))),
              showlegend=False),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(x=X5[:,0], y=X5[:,1],
              mode='markers', marker=dict(color=list(map(SetColor, y5))),
              showlegend=False),
    row=2, col=2
)

fig.update_layout(title_text='ADASYN with K_neighbors')

fig.show()
```

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, chi2, f_classif, mutual_info_classif
x, y=load_digits(return_X_y=True)
from sklearn.pipeline import make_pipeline
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
sb_chi=SelectKBest(score_func=chi2)
x_chi=sb_chi.fit_transform(x,y)
np.where(sb_chi.get_support()==1)
```

```python
sb_chi.scores_.argsort()
```

```python
sb_f=SelectKBest(score_func=f_classif)
x_f=sb_f.fit_transform(x,y)
print(sb_f.get_support().sum())
np.where(sb_f.get_support()==1)
```

```python
sb_chi.pvalues_.argsort()
```

```python
sb_f.scores_
```

```python
sb_f.pvalues_
```

```python
md1 = make_pipeline(SelectKBest(k=30), PCA(n_components=3))
x_new=md1.fit_transform(x,y)
```

```python
sb_m=SelectKBest(score_func=mutual_info_classif)
x_m=sb_m.fit_transform(x,y)
print(sb_m.get_support().sum())
np.where(sb_m.get_support()==1)
```
